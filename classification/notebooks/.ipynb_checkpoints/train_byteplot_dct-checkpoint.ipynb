# Train byteplot AND dct with resnet

```python
import pandas as pd
from fastai.basics import *
from fastai.callback.all import *
from fastai.vision.all import *
#rom koila import lazy
import numpy as np
import torch
import os
```

```python
# Check if gpu
assert(torch.cuda.is_available())
```

## Load images

```python
path = 'categorized_byteplot_and_dct'
dls = ImageDataLoaders.from_folder(path, valid="val", train="train", seed=696969, bs=32)
```

```python
learner = vision_learner(dls, resnet50)
learner.lr_find()
```

```python
learner.fit_one_cycle(1)
```

```python
learner.show_results(max_n=16)
```

```python
tta = learner.tta(use_max=True)
```

```python
interp = ClassificationInterpretation.from_learner(learner)
losses,idxs = interp.top_losses()
len(dls.valid_ds)==len(losses)==len(idxs)
interp.plot_confusion_matrix(figsize=(7,7))
```

```python
torch.save(learner, "byteplot_and_dct.pkl")
```

```python
learner = torch.load("byteplot_and_dct.pkl")
```

```python
test_images = get_image_files(f'{path}/test')
test_dl = dls.test_dl(test_images, with_labels=True)
test_dl.show_batch(max_n=6)
```

```python
preds, y = learner.get_preds(dl=test_dl)
```
